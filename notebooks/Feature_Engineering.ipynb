```python

# Step 5: Define a custom function to calculate mode
def custom_mode(x):
    modes = x.mode()
    if not modes.empty:
        return modes.iloc[0]
    else:
        return np.nan  # or return a default value or string indicating no mode
```

```python

# Step 6: Group by and aggregate
grouped = merged_df.groupby(['truck_id', 'route_id'], as_index=False).agg(
    route_avg_temp=('temp', 'mean'),
    route_avg_wind_speed=('wind_speed', 'mean'),
    route_avg_precip=('precip', 'mean'),
    route_avg_humidity=('humidity', 'mean'),
    route_avg_visibility=('visibility', 'mean'),
    route_avg_pressure=('pressure', 'mean'),
    route_description=('description', custom_mode)
)
```
